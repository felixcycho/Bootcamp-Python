```python
import pandas as pd

series = pd.Series([2, 8, -2, 10])
print(type(series))                # <class 'pandas.core.series.Series'>
print(series)
# 0     2
# 1     8
# 2    -2
# 3    10
# dtype: int64

dictionary = {
  "name" : "Leo",
  "age" : 23
}
print(type(dictionary))       # <class 'dict'>
print(dictionary)             # {'name': 'Leo', 'age': 23}

# Convert Dictionary --> DataFrame
df = pd.DataFrame(dictionary, index = [0])
print(type(df))               # <class 'pandas.core.frame.DataFrame'>
print(df)
#   name  age
# 0  Leo   23


teams = {
  "names" : ['Leo', 'Jason', 'Jennie'],
  "ages" : [23, 18, 9],
  "genders" : ['M', 'M', 'F']
}
print(type(teams))             # <class 'dict'>
print(teams)                   # {'names': ['Leo', 'Jason', 'Jennie'], 'ages': [23, 18, 9]}

df2 = pd.DataFrame(teams)
print(type(df2))               # <class 'pandas.core.frame.DataFrame'>
print(df2)
#     names  ages
# 0     Leo    23
# 1   Jason    18
# 2  Jennie     9
```

```python
df3 = pd.read_csv(titanic_url)

print(type(df3["Age"]))      # <class 'pandas.core.series.Series>

# Series, support statistics
print(df3["Age"].mean())        # 29.69911764705882
print(df3["Age"].max())         # 80.0
print(df3["Age"].min())         # 0.42

result = df3.groupby("Survived")["Age"].mean()
print(type(result))          # <class 'pandas.core.series.Series'>
print(result)
# Survived
# 0    30.626179
# 1    28.343690
# Name: Age, dtype: float64
```

```python
print(df3["Age"].isnull().any())                # True
print(df3["Name"].isnull().any())               # False

print(df3.columns[df3.isnull().any()])          # Index(['Age', 'Cabin', 'Embarked'], dtype='object')

# ETL (extract, transform, load)
# clean all data with null value
df_drop_na = df3.dropna(subset = ["Age"])
print(len(df3))                                            # 891
print(len(df_drop_na))                                     # 714

df_drop_na = df_drop_na.dropna(subset = ["Cabin"])
df_drop_na = df_drop_na.dropna(subset = ["Embarked"])
print(len(df_drop_na))                                     # 183
```

```python
# Refill null value
df_for_refill = df3.copy()
df_for_refill["Age"] = df_for_refill["Age"].fillna("N/A")
df_for_refill["Cabin"] = df_for_refill["Cabin"].fillna("N/A")
df_for_refill["Embarked"] = df_for_refill["Embarked"].fillna("N/A")

print(df_for_refill.columns[df_for_refill.isnull().any()])      # Index([], dtype='object)
print(len(df_for_refill))                                       # 891
print(df3)                                                      # 891
```

```python
# dataframe to csv
df_for_refill.to_csv("titanic_after_refill.csv")
```
